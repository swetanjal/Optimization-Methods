# Question 1: Spectral Clustering

Implement spectral clustering and evaluate on the given concentric circles dataset for this question.

## Part 1: Implementation
Implement spectral clustering function from scratch (for two clusters), taking as input the dataset. It must return the predicted clustering. Assume that the graph constructed is a fully connected graph. Use the normalized graph laplacian for this case.



```python
# Part 1: Spectral Clustering
def similarity(a, b, sigma_square):
    return np.exp(-(np.linalg.norm(a - b)**2) / sigma_square)

def spectral_clustering(X):
  """Args:
    X: numpy.array [num_samples, input_dim]
    
    Returns:
    Y_pred: numpy.array [num_samples]
      array of cluster labels
  """
  # Your code here
  D = np.zeros((X.shape[0], X.shape[0]))
  W = np.zeros((X.shape[0], X.shape[0]))
  # Compute W matrix
  for i in range(X.shape[0]):
    for j in range(X.shape[0]):
        W[i][j] = similarity(X[i], X[j], 0.01)
  # Compute degree matrix
  for i in range(X.shape[0]):
    for j in range(X.shape[0]):
        D[i][i] += W[i][j]
  # Construct normalised Laplacian
  L = np.eye(X.shape[0]) - np.sqrt(np.linalg.inv(D)) @ W @ np.sqrt(np.linalg.inv(D))
  w, v = np.linalg.eig(L)
  # Sorting eigenvectors/values in increasing order
  idx = w.argsort()   
  eigenValues = w[idx]
  eigenVectors = v[:,idx]
  # Take the second smallest eigen vector
  Y_pred = eigenVectors[:, 1].real
  # Assign cluster based on sign
  for i in range(Y_pred.shape[0]):
        if Y_pred[i] > 0:
            Y_pred[i] = 1
        else:
            Y_pred[i] = 0
  return Y_pred
```

# Question 2: Manifold Visualization
Implement the various manifold learning methods and visualize the given datasets.


## Part 1: MDS
Implement Multi-Dimensional Scaling

```python
# Part 1: MDS
def distance(a, b):
    return np.linalg.norm(a - b)
def MDS(X, k, output_dim=2):
  """Args:
    X: numpy.array [n_samples, input_dim]
    k: number of nearest neighbours to construct the knn graph
    output_dim: dimension of output data

    Returns:
    Y: numpy.array [n_samples, output_dim]
  """
  # Your code here
  dst_mtx = np.zeros((X.shape[0], X.shape[0]))
  for i in range(X.shape[0]):
    for j in range(X.shape[0]):
        dst_mtx[i][j] = distance(X[i], X[j])
  e = np.ones((X.shape[0], 1))
  H = np.eye(X.shape[0]) - (1/X.shape[0]) * e.dot(e.T)
  kernel_mtx = ((-0.5 * H).dot(dst_mtx)).dot(H)
  w, v = np.linalg.eig(kernel_mtx)
  # Sorting eigenvectors/values in increasing order
  idx = w.argsort()   
  eigenValues = w[idx]
  eigenVectors = v[:,idx]
  stk = []
  for i in range(1, output_dim + 1):
    stk.append(np.sqrt(eigenValues[-i]) * eigenVectors[:, -i])
  Y = np.column_stack(tuple(stk))
  return Y
```

## Part 2: LLE
Implement Locally Linear Embedding function

```python
# Part 2: LLE
def LLE(X, k, output_dim=2):
  """Args:
    X: numpy.array [n_samples, input_dim]
    k: number of nearest neighbours to construct the knn graph
    output_dim: dimension of output data

    Returns:
    Y: numpy.array [n_samples, output_dim]
  """
  # Your code here
  dst_mtx = kneighbors_graph(X, k, mode='connectivity', include_self=False)
  dst_mtx = dst_mtx.toarray()
  w = np.zeros((X.shape[0], k))
  for i in range(X.shape[0]):
    x_hat = []
    for j in range(k):
        x_hat.append(X[i])
    x_hat = np.column_stack(tuple(x_hat))
    N = []
    for j in range(X.shape[0]):
        if dst_mtx[i][j] != 0:
            N.append(X[j])
    N = np.column_stack(tuple(N))
    G = (x_hat - N).T @ (x_hat - N)
    # sig2 = (numpy.linalg.svd(x_hat - N,compute_uv=0))**2
    # sig2 /= sig2.sum()
    # G.flat[::k+1] += np.sum(sig2[output_dim:])
    # ll = np.linalg.solve(G, np.ones((k,1)))
    ll = np.linalg.pinv(G) @ np.ones((k, 1))
    ll = ll / np.sum(ll)
    for m in range(ll.shape[0]):
        w[i][m] = ll[m]
  I_grand = []
  W_grand = []
  for i in range(X.shape[0]):
    I = np.zeros((X.shape[0], 1))
    I[i] = 1
    I_grand.append(I)
    W = np.zeros((X.shape[0], 1))
    c = 0
    for j in range(X.shape[0]):
        if dst_mtx[i][j] != 0:
            W[j][0] = w[i][c]
            c += 1
    W_grand.append(W)
  W_grand = np.column_stack(tuple(W_grand))
  I_grand = np.column_stack(tuple(I_grand))
  M = (I_grand - W_grand) @ (I_grand - W_grand).T
  w, v = np.linalg.eig(M)
  # Sorting eigenvectors/values in increasing order
  idx = w.argsort()   
  eigenValues = w[idx].real
  eigenVectors = v[:,idx].real
  stk = []
  for i in range(1, output_dim + 1):
    stk.append(eigenVectors[:, i])
  Y = np.column_stack(tuple(stk))
  return Y
```

## Part 3: ISOMAP
Implement Isomap Visualization  

```python
# Part 3: Isomap
def ISOMAP(X, k, output_dim=2):
  """Args:
    X: numpy.array [n_samples, input_dim]
    k: number of nearest neighbours to construct the knn graph
    output_dim: dimension of output data

    Returns:
    Y: numpy.array [n_samples, output_dim]
  """
  # Your code here
  dst_mtx = kneighbors_graph(X, k, mode='distance', include_self=True)
  dst_mtx = dst_mtx.toarray()
  dst_mtx = graph_shortest_path(dst_mtx, directed = False)
  dst_mtx = dst_mtx * dst_mtx
  e = np.ones((X.shape[0], 1))
  H = np.eye(X.shape[0]) - (1/X.shape[0]) * e.dot(e.T)
  kernel_mtx = ((-0.5 * H).dot(dst_mtx)).dot(H)
  w, v = np.linalg.eig(kernel_mtx)
  # Sorting eigenvectors/values in increasing order
  idx = w.argsort()   
  eigenValues = w[idx]
  eigenVectors = v[:,idx]
  stk = []
  for i in range(1, output_dim + 1):
    stk.append(np.sqrt(eigenValues[-i]) * eigenVectors[:, -i])
  Y = np.column_stack(tuple(stk))
  return Y
```

# Question 4: Classification

Perform classification using a machine learning algorithm of your choice. Use 6k images from CIFAR-10 dataset.(5k images for training and 1k images for testing.)


*   Do dimensionality reduction on the dataset using PCA and ISOMAP.
*   Apply the classification algorithm.
*   Compare the results by changing the dimensionality of the data.
*   Use F1-score as metric.
*   Approach: Reduce the dimensionality into any two dimensions(of your choice) which are less than the initial dimensionality of the data using PCA and ISOMAP. Compare the performance metrics(F1-score) for the low dimensional data.



```python
# Code for loading CIFAR-10 dataset.
from keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X_train = X_train[:5000].reshape([5000,32*32*3])
y_train = y_train[:5000]
X_test = X_test[:1000].reshape([1000,32*32*3])
y_test = y_test[:1000]
# Initial dimensionality/number of features (32*32*3) = 3072.
```

```python
from sklearn.decomposition import PCA
def reductionUsingPCA(X, k, output_dim):
  """Args:
    X: numpy.array [n_samples, input_dim]
    k: number of nearest neighbours to construct the knn graph
    output_dim: dimension of output data

    Returns:
    pca_X: numpy.array [n_samples, output_dim]
  """
  # Enter your code here
  pca = PCA(n_components=output_dim)
  pca_X = pca.fit_transform(X)
  return pca_X
```

```python
from sklearn.manifold import Isomap

# Define the model and train

def reductionUsingISOMAP(X, k, output_dim):
  """Args:
    X: numpy.array [n_samples, input_dim]
    k: number of nearest neighbours to construct the knn graph
    output_dim: dimension of output data

    Returns:
    isomap_X: numpy.array [n_samples, output_dim]
  """
  # Enter your code here
  model = Isomap(n_neighbors=k, n_components=output_dim)
  out = model.fit_transform(X)
  return out
```

```python
# Classification Algorithm 
# Extra functions here
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
def training(X_train, y_train):
    # clf = SVC()
    clf = MLPClassifier(hidden_layer_sizes=(1024, 512, 512), max_iter=1000)
    clf.fit(X_train, y_train)
    return clf

def classification(X, n_classes, CLF):
  """Args:
    X: numpy.array [n_samples, input_dim]
    n_classes: number of classes

    Returns:
    labels: numpy_array [n_samples]
  """
  # Your code here
  labels = CLF.predict(X)
  return labels
```

```python
from sklearn.metrics import f1_score
def F1_score(y_true, y_pred):
  """Args:
    y_true: numpy.array [n_samples] , ground truth value
    y_pred: numpy.array [n_samples] , predicted value by classifier
    
    Returns:
    score: float, f1-score
  """
  # Enter your code here
  score = f1_score(y_true, y_pred, average='weighted')
  return score
```

```python
X = np.concatenate((X_train, X_test), axis = 0)
X_pca = reductionUsingPCA(X, 500, 500)
X_train_pca = X_pca[0: X_train.shape[0], :]
X_test_pca = X_pca[X_train.shape[0]: X_train.shape[0] + X_test.shape[0], :]
y_train_vectorised = []
y_test_vectorised = []
for i in range(len(y_train)):
    y_train_vectorised.append(y_train[i][0])
for i in range(len(y_test)):
    y_test_vectorised.append(y_test[i][0])
y_test_vectorised = np.array(y_test_vectorised)
y_train_vectorised = np.array(y_train_vectorised)
clf = training(X_train_pca, y_train_vectorised)
pred = classification(X_test_pca, 10, clf)
acc = 0
for i in range(len(pred)):
    if pred[i] == y_test_vectorised[i]:
        acc += 1
print("Accuracy using PCA with 500 components: ", (acc * 100/ len(pred)), "%")
print("F1 score using PCA with 500 components: ", F1_score(y_test_vectorised, pred))
```

```python
X = np.concatenate((X_train, X_test), axis = 0)
X_isomap = reductionUsingISOMAP(X, 50, 100)
X_train_isomap = X_isomap[0: X_train.shape[0], :]
X_test_isomap = X_isomap[X_train.shape[0]: X_train.shape[0] + X_test.shape[0], :]
y_train_vectorised = []
y_test_vectorised = []
for i in range(len(y_train)):
    y_train_vectorised.append(y_train[i][0])
for i in range(len(y_test)):
    y_test_vectorised.append(y_test[i][0])
y_test_vectorised = np.array(y_test_vectorised)
y_train_vectorised = np.array(y_train_vectorised)
clf = training(X_train_isomap, y_train_vectorised)
pred = classification(X_test_isomap, 10, clf)
acc = 0
for i in range(len(pred)):
    if pred[i] == y_test_vectorised[i]:
        acc += 1
print("Accuracy using Isomap with 100 components: ", (acc * 100/ len(pred)), "%")
print("F1 score using Isomap with 100 components: ", F1_score(y_test_vectorised, pred))
```
